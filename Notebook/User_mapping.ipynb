```python
import math
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
tags = pd.read_csv("../input/Tags_Filtered.csv", encoding='latin1')
#answers=pd.read_csv("Answers_Filtered.csv", encoding="latin1")
questions=pd.read_csv("../input/Questions_Filtered.csv",encoding="latin1")
```

```python
user_dict = dict()
for index, row in questions.iterrows():
    if row['OwnerUserId'] not in user_dict:
        tag_score_dict=dict()
        tag_count_dict=dict()
        temp_rows=tags[tags['Id'] == row['Id']]
        
        for index1,rows1 in temp_rows.iterrows():
            if rows1['Tag'] not in tag_count_dict:
                tag_count_dict[rows1['Tag']] = 1
                tag_score_dict[rows1['Tag']] = max(row['Score']//len(temp_rows),1)
            else:
                tag_count_dict[rows1['Tag']] +=1
                tag_score_dict[rows1['Tag']] += max(row['Score']//len(temp_rows),1)
        user_dict[row['OwnerUserId']]=(tag_count_dict,tag_score_dict)
    
    else:
        temp_rows=tags[tags.Id == row['Id']]
        tag_count_dict = user_dict[row['OwnerUserId']][0]
        tag_score_dict = user_dict[row['OwnerUserId']][1]
        for index1,rows1 in temp_rows.iterrows():
            if rows1['Tag'] not in tag_count_dict:
                    tag_count_dict[rows1['Tag']] = 1
                    tag_score_dict[rows1['Tag']] = max(row['Score']//len(temp_rows),1)
            else:
                tag_count_dict[rows1['Tag']]+= 1
                tag_score_dict[rows1['Tag']]+= max(row['Score']//len(temp_rows),1)
    
```
